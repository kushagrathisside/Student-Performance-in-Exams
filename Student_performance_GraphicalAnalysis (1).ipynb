```python
data['av_score']=(data['math score']+data['reading score']+data['writing score'])/3
```

```python
sns.scatterplot(x=data['av_score'],y=data['math score'],hue=data['race/ethnicity'])
```

```python
sns.scatterplot(x=data['av_score'],y=data['math score'],hue=data['gender'])
```

```python
plt.hist(data['av_score'])
```

```python
for i in range(0, 1000):
    data['av_score'][i] = ceil(data['av_score'][i])

plt.rcParams['figure.figsize'] = (15,8)
sns.distplot(data['av_score'], color = 'red')

plt.xlabel('Average_score')
plt.ylabel('Count')
plt.show()
```

```python
m=data['math score']
r=data['reading score']
w=data['writing score']

plt.rcParams["figure.figsize"] = [15, 10]
plt.rcParams["figure.autolayout"] = True

data = {'MATH': m, 'READ': r,'WRITE': w}
fig, ax = plt.subplots()

ax.boxplot(data.values())
ax.set_xticklabels(data.keys())

plt.show()
```
